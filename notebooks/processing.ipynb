### Preprocesing

##### Import Stopword Indonesia

```python
import requests
stopwords_url = "https://raw.githubusercontent.com/louisowen6/NLP_bahasa_resources/master/combined_stop_words.txt"

response = requests.get(stopwords_url)

stopwords = response.text.splitlines()

stopwords = set(stopwords)

print(list(stopwords)[:10])
```

```python
kata_negasi = ['tidak', 'bukan', 'ga', 'gk', 'jangan']
for kata in kata_negasi:
    if kata in stopwords:
        stopwords.remove(kata)
```

##### Import Slang Indonesia

```python
import json

slang_url = "https://raw.githubusercontent.com/louisowen6/NLP_bahasa_resources/master/combined_slang_words.txt"

response = requests.get(slang_url)

kamus_slang_custom = response.json()

kamus_slang_custom.update({
    'tbtb': 'tiba tiba',
    'gk': 'tidak',
    'kdng': 'kadang'
})

print("Total kata di kamus baru:", len(kamus_slang_custom))
print("Contoh isi kamus (tbtb):", kamus_slang_custom['tbtb'])
```

#### Cleaning

```python
def clean_string(text):
    text = str(text)

    # 1. Remove Metadata & Footer Removal
    text = text.strip()

    # 2. Remove URL
    text = re.sub(r'https?://\S+|www\.\S+', ' ', text)

    # 3. Lowercase
    text = text.lower()

    # 4. Remove Numbers
    text = re.sub(r'\d+', ' ', text)

    # 5. Remove Punctuation (Menghapus tanda baca & simbol/emoji)
    text = re.sub(r'[^a-z\s]', ' ', text)

    # 6. Merapikan spasi berlebih hasil pembersihan
    text = ' '.join(text.split())

    return text
```

```python
# Pembersihan Berbasis Kata (Tokens)

def process_tokens(cleaned_sentence):
    text = str(cleaned_sentence).lower()
    # Tokenization
    tokens = text.split()

    final_tokens = []
    for token in tokens:
        token = token.strip()

        # Normalisasi Slang
        if token in kamus_slang_custom:
            token = kamus_slang_custom[token]

        # Stopword Removal
        if token not in stopwords:
            final_tokens.append(token)

    # Mengembalikan hasil akhir dalam bentuk kalimat bersih atau list token
    return ' '.join(final_tokens)
```

```python
# Membersihkan String
df_gojek_test['step1_string'] = df_gojek_test['content'].apply(clean_string)

# Olah token dan normalisasinya
df_gojek_test['clean_content'] = df_gojek_test['step1_string'].apply(process_tokens)

# Perbandingannya
print(df_gojek_test[['content', 'clean_content']].head(100))
```

#### Labeling

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
from deep_translator import GoogleTranslator
from tqdm import tqdm

# Inisialisasi SentimentIntensityAnalyzer dan GoogleTranslator
analyzer = SentimentIntensityAnalyzer()
translator = GoogleTranslator(source='id', target='en')

tanyain_teks = df_gojek_test['clean_content'].fillna("").astype(str).tolist()

hasil_terjemahan = []
batch_size = 100

print("Memulai proses translasi berbasis batch...")
for i in tqdm(range(0, len(tanyain_teks), batch_size)):
    batch = tanyain_teks[i:i+batch_size]
    try:
        # Masukkan list berisi 100 teks, hasil outputnya juga list berisi 100 terjemahan
        translated_batch = translator.translate_batch(batch)
        hasil_terjemahan.extend(translated_batch)
    except Exception as e:
        # Jaring pengaman jika satu batch gagal, isi dengan teks kosong agar ukuran list tetap singkron
        hasil_terjemahan.extend([""] * len(batch))
```

```python
def hitung_vader_dari_en(teks_en):
    if not teks_en or teks_en.strip() == "":
        return 'netral'
    scores = analyzer.polarity_scores(teks_en)
    compound = scores['compound']
    if compound >= 0.05: return 'positif'
    elif compound <= -0.05: return 'negatif'
    else: return 'netral'

# Petakan hasil terjemahan langsung ke kolom sentimen
df_gojek_test['sentimen'] = [hitung_vader_dari_en(t) for t in hasil_terjemahan]
```

```python
df_gojek_test[['content', 'clean_content', 'sentimen']].head(100)
```

```python
df_gojek_test['sentimen'].value_counts()
```

```python
df_gojek_test.to_csv('gojek_review_bersih.csv', index=False)
```
